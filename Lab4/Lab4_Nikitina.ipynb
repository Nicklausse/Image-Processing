## Определение возраста дерева

```python
img_tree = imread('11.jpg')
img_tree = rgb2gray(img_tree)
imshow(img_tree)
```

```python
img_tree_cut = img_tree[1610:1650, 285:1490]

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut, cmap ='gray')
```

```python
img_tree_cut_eq = equalize(img_tree_cut, selem=disk(30))

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut_eq, cmap ='gray')
```

Эквализация гистограммы необходима вследствие того, что светлые кольца слева выглядят так же, как темные кольца.

```python
img_tree_cut_eq_float = img_as_float(img_tree_cut_eq>170)

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut_eq_float, cmap ='gray')
```

```python
img_tree_cut_eq_float_dil = dilation(img_tree_cut_eq_float, rectangle(4,3))

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut_eq_float_dil, cmap ='gray')
```

```python
line = np.mean(img_tree_cut_eq_float_dil, axis = 0)
img_tree_cut_eq_float_dil = np.array([line]*40)

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut_eq_float_dil, cmap ='gray')
```

```python
img_tree_cut_eq_float_dil = img_as_float(binary_erosion(img_tree_cut_eq_float_dil, rectangle(2,2)))

plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut_eq_float_dil, cmap ='gray')
```

```python
def get_tree_age(X): 
    count_series = 0
    
    for i in range(len(X)):
        if X[i]>0.3:
            X[i]=1
        else:
            X[i]=0
    
    for i in range(len(X)-1):
        if X[i+1]!=X[i]: #если следующее значение совпадает с текущим
            if X[i]==0:
                count_series+=1 #количество серий
    print('Возраст дерева:', count_series)
```
